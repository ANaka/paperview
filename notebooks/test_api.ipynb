```python


# Set the base URL and format
base_url = "https://api.biorxiv.org/details/"
format = "json"

# Set the server and interval or DOI
server = "biorxiv"
interval = "2018-08-21/2018-08-28"

# Set the cursor and format
cursor = 45
format = "json"

# Construct the URL
url = f"{base_url}{server}/{interval}/{cursor}/{format}"

# Make the request
response = requests.get(url)

# Access the metadata elements
data = response.json()
doi = data['doi']
title = data['title']
authors = data['authors']
author_corresponding = data['author_corresponding']
author_corresponding_institution = data['author_corresponding_institution']
date = data['date']
version = data['version']
category = data['category']
jats_xml_path = data['jats_xml_path']
abstract = data['abstract']
published = data['published']
```
